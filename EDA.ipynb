```python
df=df[['coaster_name',
    #'Length', 'Speed',
    'Location',
    'Status',
    #'Opening date',
      # 'Type',
     'Manufacturer',
    #'Height restriction', 'Model', 'Height',
      # 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date',
    #'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced',
     'latitude',
     'longitude',
     'Type_Main',
       'opening_date_clean',
    # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    # 'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df=df.rename(columns={'coaster_name':'Coaster_name',
                   'year_introduced':'Year_introduced',
                   'latitude':'Latitude',
                   'longitude':'Longitude',
                   'opening_date_clean':'Opening_date',
                   'speed_mph':'Speed_mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'
                   })
```

**Step 3: Feature Understanding**

(Univariate analysis)


*   Plotting Feature Distributions
*   Histogram
*   KDE
*   Boxplot

```python
ax=df['Year_introduced'].value_counts().head(10).plot(kind='bar',title='Top Years Coasters Introduced')
ax.set_xlabel('Year_introduced')
ax.set_ylabel('Count')
```

```python
ax=df['Speed_mph'].plot(kind='hist',bins=20,title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
```

```python
ax=df['Speed_mph'].plot(kind='kde',title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.show()
```

```python
df.plot(kind='scatter',x='Speed_mph',y='Height_ft',title='Coaster Speed vs Height')
plt.show()
```

```python
ax=sns.scatterplot(data=df,x='Speed_mph',y='Height_ft',hue='Year_introduced')
ax.set_title('Coaster Speed vs Height')
ax.set_xlabel('Speed (mph)')
ax.set_ylabel('Height (ft)')
plt.show()
```

```python
sns.pairplot(df,vars=['Speed_mph','Height_ft','Year_introduced','Gforce'],hue='Type_Main')
plt.show()
```

```python
df_corr=df[['Year_introduced','Speed_mph',
    'Height_ft','Gforce']].corr()
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
```

**Step 5: Ask a Question about the Data**

Try to answer a question you have about the data using a plot or statistic.

What are de the locations with the fastest roller coasters (minimun of 10)?

```python
ax=df.query('Location!="Other"') \
.groupby('Location') \
['Speed_mph'].agg(['mean','count']) \
.query('count>=10') \
.sort_values('mean') \
['mean'].plot(kind='barh',title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed (mph)')
```
